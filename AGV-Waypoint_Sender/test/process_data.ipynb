```python
file = "/home/jakub/Documents/GitHub/Master-s-Degree/results.txt"
results = {}
with open(file, "r") as f:
    for line in f.readlines():
        line = line[1:-2].replace("' ", "").replace(" '", "").replace("'", "")
        line = line.split(",")
        results[line[0]] = [float(line[0].split(" ")[1]),float(line[0].split(" ")[1])*2,float(line[0].split(" ")[2]),float(line[0].split(" ")[3]),int(line[-1]),*[int(i) for i in line[1:-1]]]

        # results[line[0]] = [int(line[-1]),line[0].split(" ")[1],line[0].split(" ")[2],line[0].split(" ")[3],line[0].split(" ")[4]]
        # results[line[0]] = [int(i) for i in line[1:]]
import pandas as pd

df = pd.DataFrame.from_dict(results, orient="index", columns=["Kp","Ka","Lookahead","Rotation", "Total","P1", "P2", "P3", "P4", "P5", "P6"])
df["Total"] = df["Total"] / 10**9
df["P1"] = df["P1"] / 10**9
df["P2"] = df["P2"] / 10**9
df["P3"] = df["P3"] / 10**9
df["P4"] = df["P4"] / 10**9
df["P5"] = df["P5"] / 10**9
df["P6"] = df["P6"] / 10**9
# df = pd.DataFrame.from_dict(results, orient="index", columns=["Kp","Ka","Lookahead","Rotation","P1", "P2", "P3", "P4", "P5", "P6", "Total"])
# df.sort_values(by=['P1'], inplace=True)
df.sort_values(by=['Total'], inplace=True)

df.to_csv("../data/results.csv")
df
```
